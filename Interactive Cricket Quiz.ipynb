```python
# ==========================================
# 🏏 CRICKET QUIZ APPLICATION
# PART 1: Setup + Questions
# ==========================================

print("==========================================")
print("🏏 WELCOME TO CRICKET QUIZ APPLICATION 🏏")
print("==========================================")

name = input("\nEnter Your Name: ")

print("\nHello", name + "!")
print("Test your cricket knowledge 🏏")


# =========================
# Difficulty Selection
# =========================

print("\nSelect Difficulty Level:")
print("1. Easy")
print("2. Medium")
print("3. Hard")

level = input("Enter your choice (1/2/3): ")


# =========================
# Easy Questions
# =========================

easy_questions = [

{"question":"Who is known as Sultan of Swing?",
"options":["a. Wasim Akram","b. Shoaib Akhtar","c. Waqar Younis"],
"answer":"a"},

{"question":"How many players are there in a cricket team?",
"options":["a. 9","b. 10","c. 11"],
"answer":"c"},

{"question":"Which country won the 1992 World Cup?",
"options":["a. India","b. Pakistan","c. Australia"],
"answer":"b"},

{"question":"How many balls are in one over?",
"options":["a. 5","b. 6","c. 8"],
"answer":"b"},

{"question":"Which format has 20 overs?",
"options":["a. ODI","b. Test","c. T20"],
"answer":"c"}

]


# =========================
# Medium Questions
# =========================

medium_questions = [

{"question":"Who has scored most ODI centuries?",
"options":["a. Virat Kohli","b. Sachin Tendulkar","c. Babar Azam"],
"answer":"a"},

{"question":"Pakistan won Champions Trophy in which year?",
"options":["a. 2015","b. 2017","c. 2019"],
"answer":"b"},

{"question":"Who won T20 World Cup 2022?",
"options":["a. India","b. Pakistan","c. England"],
"answer":"c"},

{"question":"Who is famous for helicopter shot?",
"options":["a. Dhoni","b. Kohli","c. Babar"],
"answer":"a"},

{"question":"Pakistan captain in 1992 World Cup?",
"options":["a. Imran Khan","b. Wasim Akram","c. Inzamam"],
"answer":"a"}

]


# =========================
# Hard Questions
# =========================

hard_questions = [

{"question":"Highest wicket taker in T20I cricket?",
"options":["a. Rashid Khan","b. Tim Southee","c. Shakib"],
"answer":"b"},

{"question":"Who has most international centuries?",
"options":["a. Sachin","b. Kohli","c. Ponting"],
"answer":"a"},

{"question":"Who took 800+ Test wickets?",
"options":["a. Warne","b. Muralitharan","c. Anderson"],
"answer":"b"},

{"question":"Who won first Cricket World Cup?",
"options":["a. West Indies","b. Australia","c. England"],
"answer":"a"},

{"question":"Champions Trophy 2025 hosted by?",
"options":["a. India","b. Pakistan","c. England"],
"answer":"b"}

]


# =========================
# Select Questions
# =========================

if level == "1":
    questions = easy_questions
    difficulty = "Easy"

elif level == "2":
    questions = medium_questions
    difficulty = "Medium"

elif level == "3":
    questions = hard_questions
    difficulty = "Hard"

else:
    questions = easy_questions
    difficulty = "Easy"


print("\nSelected Difficulty:", difficulty)
print("Total Questions:", len(questions))
print("Quiz is ready! 🏏")
# ==========================================
# PART 2: Quiz Logic + 10 Seconds Timer
# ==========================================

import time


score = 0


# ==========================================
# Start Quiz
# ==========================================

for index, q in enumerate(questions):

    print("\n==========================================")
    print(f"Question {index + 1}/{len(questions)}")
    print("==========================================")

    print(q["question"])
    print()

    for option in q["options"]:
        print(option)


    start_time = time.time()

    answer = input("\nEnter your answer (a/b/c): ")

    end_time = time.time()


    total_time = end_time - start_time


    # Time Check

    if total_time > 10:

        print("\n⏰ Time Over!")
        print("No marks awarded.")

    else:

        if answer.lower() == q["answer"]:

            print("\n✅ Correct Answer!")
            score += 1

        else:

            print("\n❌ Wrong Answer!")

            for option in q["options"]:

                if option.startswith(q["answer"]):
                    print("Correct Answer:", option)


    print("Current Score:", score)


print("\n==========================================")
print("Quiz Questions Completed 🏏")
print("==========================================")
# ==========================================
# PART 3: Final Result + Leaderboard
# ==========================================


print("\n==========================================")
print("🏏 QUIZ COMPLETED")
print("==========================================")


print("Player Name:", name)
print("Difficulty Level:", difficulty)

print("Final Score:", score, "/", len(questions))


percentage = (score / len(questions)) * 100

print("Percentage:", percentage, "%")


# =========================
# Performance Message
# =========================

if percentage >= 90:

    print("🏆 Outstanding! You are a Cricket Expert!")

elif percentage >= 70:

    print("🌟 Excellent! Great Cricket Knowledge!")

elif percentage >= 50:

    print("👍 Good Job! Keep Improving!")

else:

    print("😊 Nice Try! Practice More!")



# =========================
# Leaderboard Save
# =========================

record = (
    f"{name} | "
    f"{difficulty} | "
    f"{score}/{len(questions)} | "
    f"{percentage}%\n"
)


with open("leaderboard.txt", "a") as file:

    file.write(record)



# =========================
# Display Leaderboard
# =========================

print("\n==========================================")
print("🏆 LEADERBOARD")
print("==========================================")


with open("leaderboard.txt", "r") as file:

    players = file.readlines()


for number, player in enumerate(players, start=1):

    print(number, ".", player.strip())


print("\nThank you for playing Cricket Quiz 🏏")
```
